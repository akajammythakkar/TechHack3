```python
from roboflow import Roboflow
from google.colab import userdata
api_key = userdata.get('RoboflowAPI')
rf = Roboflow(api_key=api_key)
project = rf.workspace("airborne-object-detection").project("airborne-object-detection-4-aod4")
version = project.version(6)
dataset = version.download("yolov8")
```

```python
!yolo detect train data="/content/Airborne-Object-Detection-4-AOD4-6/data.yaml" model=yolov8n.pt epochs=1 freeze=10 save=True name=test optimizer='Adam' batch=64 lr0=0.003 plots=True
```
